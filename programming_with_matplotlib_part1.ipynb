# Programming with matplotlib (Part 1)

```python
import numpy as np
from matplotlib import pyplot as plt
# or ( import matplotlib.pyplot as plt )
```

## Option1: Statefull Approach

```python
# (plot 1: a piecewise linear graph, where x and y are manually defined lists. )
x = [1, 2, 3, 4, 5] # x-points
y = [10, 20, 25, 30, 40] # y-points

plt.plot(x, y) # plot function connects the points(x, y) with a line
plt.show() 
```

```python
# (plot 2: a smooth curved plot, where x and y are generated  )

# np.linspace() is a NumPy function that returns evenly spaced values over a specified interval
#It’s commonly used when you want to plot a smooth curve for a function

x = np.linspace(-2, 2, 20)  #x is generated using np.linspace(start, end, num of elements), which creates 20 evenly spaced points between -2 and 2.
y = x**3 # y is generated by appling a mathematical function to each value of x, so it’s a smooth cubic curve.

plt.plot(x,y) # will plot points on the current plot/ graph
plt.show()
```

## Option 2: Stateless approach

plt.subplot()  vs  fig.add_subplot()  vs  plt.subplots() are  three different ways to create plots using Matplotlib

### a. plt.subplot()

```python
# Use this method to add an Axes to the current figure 

fig = plt.figure() # simply creates a new figure of size 640x480 with 0 Axes (and returns a Figure object, which we store as fig.)

# THIS IS THE MAIN LINE OF CODE:
ax = plt.subplot() # adds an axes (subplot) to the current figure. ( and it returns an Axes object, which we store as ax)

ax.plot(x,y) # x and y points were generated previously. Those same points r being used here.
```

plt.subplot() relies on the current figure state managed by pyplot, so it can be less explicit or harder to manage when working with multiple figures.

### b. fig.add_subplot()

```python
# Use it to add an Axes to the figure on which it is called

fig = plt.figure() # again creates a new figure ...

# (MAIN LINE OF CODE) THIS IS THE LINE THAT CHANGED: 
ax = fig.add_subplot() # explicitly adds an axes to that specific figure. ( and it returns an Axes object, which we store as ax)

ax.plot(x,y)
```

### c. plt.subplots()

```python
# Use this method to create a figure and a add set of subplots in a single call

# (MAIN LINE OF CODE) THIS IS THE LINE THAT CHANGED: 
fig, ax = plt.subplots() #Combines figure and subplot creation in one step.


ax.plot(x,y)
```

side note -> Difference b/w the two: 
1. plt.plot(x, y) will plot points on the current axes, .
2. ax.plot(x, y) will plot points on Plots on a specific Axes object.

### How to Draw a Line Chart


```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

fig,ax = plt.subplots() #create a figure and a add a subplots in a single call

#This draws a line graph on the ax (axes) using: (years on the x-axis), (chemical_exports on the y-axis)
ax.plot(years, chemical_exports) #Internally, it connects each (x, y) pair with a straight line:
#Example: connects (2010, 0.810) to (2011, 0.831) and so on.

# plt.show()  #we dont have to call plt.show() in anaconda jupyter notebook to plot the graph. however for vs code or others, youll have to write this.
```

### Enhance the Graph Step by Step

As you can see that the above were not much efficient. So, we add labels, ticks, legends etc to make the plots more understandable and self explanatory.

#### a. Adding Chart Title

```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

fig,ax = plt.subplots()
ax.plot(years, chemical_exports)

# ( NEW LINE ):
plt.title("LCI Exports in last 12 years") # 
# In OO style, it’s better to use  ax.set_title()  instead of  plt.title().
# But ax.set_title() does not work in anaconda jupyter notebook.
```

#### b. Setting x/y-Labels

```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

fig,ax = plt.subplots()
ax.plot(years, chemical_exports)

# ( NEW LINE ):
ax.set_xlabel("Years") # give a name to the values along x axis.
# ( NEW LINE ):
ax.set_ylabel("Amount (Million US$)") # give a name to the values along y axis.


plt.title("LCI Exports in last 12 years")
```

#### c. Modifying Ticks

```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

fig,ax = plt.subplots()
ax.plot(years, chemical_exports)

ax.set_xlabel("Years")
ax.set_ylabel("Amount (Million US$)")

# ( THIS IS A SET OF NEW LINES ADDED FOR MODIFYING TICKS):
# Ticks are the values you want to show over the axes. (instead of the default auto-spacing)

# xvals = np.arange(start, end, step)
xvals = np.arange(2011, 2022, 2) # uses NumPy's arange() function to generate x-axis values from 2011 to just before 2022, spaced by 2 years.

# yvals = np.linspace(start, end, num_of_values)
yvals = np.linspace(0.80, 0.98, 10) # Uses linspace() to generate 10 evenly spaced y-axis values between 0.80 and 0.98.

ax.set_xticks(xvals) #Sets the x-axis tick locations to the values in xvals. Only those years (2011, 2013, ..., 2021) will appear as tick labels on the x-axis.
ax.set_yticks(yvals) #Sets the y-axis tick locations to the values in yvals. The y-axis will now show these specific values as tick labels.


plt.title("LCI Exports in last 12 years")
```

#### d. Adding Multiple Lines in a Graph

You can draw multiple lines on the same plot by calling ax.plot() again with a different dataset:

```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]

# ( NEW LINE ):
# Each of the following values paired with the corresponding value in years' list will form points on the graph. this will be our new (2nd) line.
medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940] 

fig,ax = plt.subplots()
ax.plot(years, chemical_exports)

# ( NEW LINE ):
ax.plot(years, medicine_exports) #This draws a line graph on the ax (axes) using: (years on the x-axis), (medicine_exports on the y-axis)

ax.set_xlabel("Years")
ax.set_ylabel("Amount (Million US$)")

xvals = np.arange(2011, 2022, 2)
yvals = np.linspace(0.80, 0.98, 10)
ax.set_xticks(xvals)
ax.set_yticks(yvals)


plt.title("LCI Exports in last 12 years")
```

#### e. Adding Chart Legend

A chart legend is a small rectangular box at the side of the chart. It contains a short sample of the line drawn next to its name. Basically, its purpose is to help identify each line drawn on the chart/plot/graph. A legend shows what each line on the plot is representing. it's like telling the name of each line.

##### Option 1: 

```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]

fig,ax = plt.subplots()
ax.plot(years, chemical_exports) # Plot line 1
ax.plot(years, medicine_exports) # Plot line 2

ax.set_xlabel("Years")
ax.set_ylabel("Amount (Million US$)")

xvals = np.arange(2011, 2022, 2)
yvals = np.linspace(0.80, 0.98, 10)
ax.set_xticks(xvals)
ax.set_yticks(yvals)


plt.title("LCI Exports in last 12 years")

# ( NEW LINE ):
ax.legend(["Chemicals", "Medicines"]) # ax.legend(["name of line1", "name of line2"])
# The order in which you plot the lines ( ax.plot(line1), ax.plot(line2) ) and the order in which u name them ( ax.legend(["line1", "line2"]) )
# should be the same. the name written first will be given to whatever line was plotted first.
# May also be written as: plt.legend(["Chemicals", "Medicines"]), but not preffered.
```

##### Option 2: 

```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]

fig,ax = plt.subplots()
ax.plot(years, chemical_exports, label="Chemicals") # naming line 1 
ax.plot(years, medicine_exports, label="Medicines") # naming line 2
# Naming lines side-by-side, saves one from the trouble of remembering the order in which lines were plotted.
# now you don't have to name them when calling the legend.

ax.set_xlabel("Years")
ax.set_ylabel("Amount (Million US$)")

xvals = np.arange(2011, 2022, 2)
yvals = np.linspace(0.80, 0.98, 10)
ax.set_xticks(xvals)
ax.set_yticks(yvals)


plt.title("LCI Exports in last 12 years")

# ( NEW LINE ):
# Now here u dont have to name the lines, u can also leave the parenthesis () empty.
ax.legend(loc='best') #This specifies the location where the legend is to be displayed.
# ax.legend(loc='upper right')  #this is the default location.
# ax.legend(loc='upper left')
# ax.legend(loc='lower right')
# ax.legend(loc='lower left')
```

### Sub-Plots

A subplot is a smaller plot (or axes) inside a larger figure.
When you want to display multiple plots side by side or in a grid layout within the same figure window, you use subplots.

```python
fig, ax = plt.subplots(nrows=3, ncols=3)
# fig, ax = plt.subplots(nrows=2, ncols=2)
# fig, ax = plt.subplots(nrows=2, ncols=1)
# fig, ax = plt.subplots(nrows=3, ncols=1)
print(ax)
```

#### a. Multiple Axes Objects within a Figure Object

```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1)

ax1.plot(years, chemical_exports, label='Chemicals', marker='o', c='b', ls='-', lw=2) 
ax1.set_title("LCI Chemical exports in last 12 years")
ax1.set_xlabel("Years")
ax1.set_ylabel("Amount (Million Rs.)")

ax2.plot(years, medicine_exports, label='Medicines', marker='x', c='r', ls=':', lw=2) 
ax2.set_title("LCI Medicine exports in last 12 years")
ax2.set_xlabel("Years")
ax2.set_ylabel("Amount (Million Rs.)")

plt.tight_layout()
```

fig: single canvas

ax1, ax2: two vertically arranged subplots (axes) on that figure

Both plots share the same figure window — useful for side-by-side or stacked comparisons

#### b. Multiple Figure Objects Having Single Axes

```python
chemical_exports = [0.810, 0.831, 0.895, 0.91, 0.915, 0.926, 0.945, 0.931, 0.919, 0.921, 0.920, 0.919]
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
medicine_exports = [0.791, 0.818, 0.832, 0.816, 0.840, 0.833, 0.835, 0.838, 0.842, 0.910, 0.930, 0.940]

fig1, ax1 = plt.subplots()
fig2, ax2 = plt.subplots()

ax1.plot(years, chemical_exports, label='Chemicals', marker='o', c='b', ls='-', lw=2) 
ax1.set_title("LCI Chemical exports in last 12 years")
ax1.set_xlabel("Years")
ax1.set_ylabel("Amount (Million Rs.)")

ax2.plot(years, medicine_exports, label='Medicines', marker='x', c='r', ls=':', lw=2) 
ax2.set_title("LCI Medicine exports in last 12 years")
ax2.set_xlabel("Years")
ax2.set_ylabel("Amount (Million Rs.)")

plt.tight_layout()
```

fig1 and fig2: two completely separate figure windows

ax1 and ax2: each is the axes of its respective figure

Each plot opens independently — they don’t share layout or style settings

You would need two separate plt.tight_layout() calls if you customize both
